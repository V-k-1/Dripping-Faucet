```python
def peaks(fx,vol,threshold,gap=1000):
    index=[]
    '''try : threshold=np.max(fx[:background])
    except : threshold=np.max(fx[background[0]:background[1]])'''
    if max(fx[:gap])>threshold : index.append(list(fx).index(max(fx[:gap]),0,gap))
    i=gap
    while i<len(fx)-gap:
        if fx[i]>threshold:
            if fx[i]==max(fx[i-gap:i+gap]): 
                index.append(i)
                i=i+gap-1
        i=i+1
    '''for i in range(gap,len(fx)-gap):
        if fx[i]>threshold:
            if fx[i]==max(fx[i-gap:i+gap]): index.append(i)'''
    if max(fx[-gap:])>threshold : index.append(list(fx).index(max(fx[-gap:]),-gap,-1))
    index_val=[fx[i] for i in index]
    plt.figure(figsize=(50,10))
    plt.plot(fx)
    plt.plot(index,index_val,'ko')
    plt.plot([index[0],index[-1]],[threshold,threshold])
    plt.show()
    return index
def delay_plot(index):
    index=np.array(index)/32000
    #print(index)
    t1,t2=[],[]
    for t in range(1,len(index)-1):
        t1.append(index[t]-index[t-1]);t2.append(index[t+1]-index[t])
    return(t1,t2)
```

```python
vols
```

```python
thresholds,gaps=[0.0075,0.01,0.01,0.01],[5000,5000,5000,1000]
avgtf2=[0,0,0,0]
for i in range(len(files)):
    avgtf2[i]=peaks(fxs[i],vols[i],thresholds[i],gaps[i])
```

```python
np.save('avgtf2-b2',avgtf2, allow_pickle=True, fix_imports=True)
```

```python
for i in range(len(avgtf2)):
    x,y=delay_plot(avgtf2[i])
    plt.plot(x,y,'o',label=FFF[i])
    plt.xlim(0,.75)
    plt.ylim(0,.75)
plt.legend()
plt.show()
```

```python
plt.figure()
Dt=[]
for index in avgtf2:
    dt=[]
    for i in range(len(index)-1):
        dt.append((index[i+1]-index[i])/32000)
    plt.hist(dt,bins=100)
    plt.show()
    Dt.append(dt)
```

```python
flow_rates=[]
wcpd,std=[],[]
for i in range(len(fxs)):
    index_val=[fxs[i][j] for j in avgtf2[i]]
    flow_rates.append(vols[i]*32000/len(fxs[i]))
    wcpd.append(vols[i]/len(avgtf2[i]))
    ratio=wcpd[i]/np.mean(index_val)
    std.append(np.std(index_val)*ratio)
plt.errorbar(flow_rates,wcpd,fmt='o',yerr=std)
plt.show()
```

```python
plt.figure(figsize=(10,5))
for i in range(len(Dt)):
    plt.plot(Dt[i],label='flow rate = '+str(flow_rates[i]))
plt.legend()
plt.show()
```

```python
FFF1=[30,41,56,25]
files1=['audio/'+ str(i)+'.wav' for i in FFF1 ]
fxs1,vols1=[],[]
```

```python
for i in range(len(FFF1)):
    sampFreq, snd = wavfile.read(files1[i])
    fx = snd[:,0]/(2.**15)
    '''plt.figure(figsize=(50,10))
    plt.plot(fx,'b')
    plt.show()'''
    fxs1.append(fx)
```

```python
plt.figure(figsize=(10,5))
for i in range(4):
    x,y=delay_plot(avgtf1[i])
    plt.scatter(x,y,marker='o',s=30,label=FFF1[i])
plt.xlim(0,1.75)
plt.ylim(0,1.75)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
for i in range(len(Dt1)):
    plt.plot(Dt1[i],label='flow rate = '+str(flow_rates1[i]))
plt.legend()
plt.show()
```
